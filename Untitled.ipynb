```python
import pandas as pd
import glob
import os

# Specify the folder where the CSV files are located
folder_path = r"C:\\Users\\RST\\Desktop\\Minor project\\Index data 1 min"

# Get a list of all CSV files in the folder
csv_files = glob.glob(os.path.join(folder_path, "*.csv"))

# Print the list of files found
print(f"Files found: {csv_files}")

# Initialize an empty list to hold each DataFrame
dataframes = []

# Loop through the list of CSV files and read each one into a DataFrame
for file in csv_files:
    print(f"Reading file: {file}")
    try:
        df = pd.read_csv(file)
        if not df.empty:
            dataframes.append(df)
        else:
            print(f"Warning: {file} is empty.")
    except Exception as e:
        print(f"Error reading {file}: {e}")

# Ensure there are DataFrames to concatenate
if dataframes:
    # Concatenate all DataFrames into a single DataFrame
    combined_df = pd.concat(dataframes, ignore_index=True)

    # Save the combined DataFrame to a new CSV file
    combined_df.to_csv(r"C:\\Users\\RST\\Desktop\\Minor project\\Index data 1 min\\combined_file.csv", index=False)

    print("Files have been combined successfully!")
else:
    print("No valid DataFrames to concatenate.")
```

```python
# Load your DataFrame (assumes CSV, adjust if needed)
df = pd.read_csv("C:\\Users\\RST\\Desktop\\Minor project\\combined_file_cleaned.csv")

# Calculate '% Change' based on the previous Close value
df['% Change'] = df['Close'].pct_change() * 100

# Calculate average close price
average_close = df['Close'].mean()

# Calculate '% Change vs Average' using a smaller difference in percentage
df['% Change vs Average'] = ((df['Close'] - average_close) / average_close) * 100

# Replace missing or NaN values for '% Change' and '% Change vs Average'
df['% Change'].fillna(0, inplace=True)
df['% Change vs Average'].fillna(0, inplace=True)

# Save the updated DataFrame
df.to_csv('combined_file_with_calculated_changes.csv', index=False)

print("Missing percentage changes calculated and file saved!")
```
